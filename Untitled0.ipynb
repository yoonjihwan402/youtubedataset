```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
from scipy.stats import pearsonr


# 인코딩 변경해서 시도
df = pd.read_csv("KR_youtube_trending_data.csv")
```

```python
df.head()
```

```python
print(df.columns)
df.columns = df.columns.str.strip()
```

```python
df = df[['trending_date', 'publishedAt', 'title', 'video_id', 'likes', 'view_count']]
```

```python
print(df.columns)
```

```python
df['trending_date'] = pd.to_datetime(df['trending_date'], errors='coerce')
df['publishedAt'] = pd.to_datetime(df['publishedAt'], errors='coerce')

# 트렌드에 오르기까지 걸린 일수
df['days_to_trend'] = (df['trending_date'] - df['publishedAt']).dt.days

# 업로드 시간대 파생 변수
df['publish_hour'] = df['publishedAt'].dt.hour
```

```python
df['title_len'] = df['title'].apply(lambda x: len(str(x)))
df['title_word_count'] = df['title'].apply(lambda x: len(str(x).split()))
```

```python
df = df.drop(['title', 'video_id', 'trending_date', 'publishedAt'], axis=1)
```

```python
X = df.drop('view_count', axis=1)
y = df['view_count']

# 로그 변환으로 스케일 안정화 (선택 사항)
import numpy as np
y_log = np.log1p(y)

# 훈련 / 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 예측
y_pred_log = model.predict(X_test)
y_pred = np.expm1(y_pred_log)  # 다시 원래 스케일로 복원
y_actual = np.expm1(y_test)

# 평가
rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
r2 = r2_score(y_actual, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.3f}")
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
sns.barplot(x=feature_importances.values, y=feature_importances.index)
plt.title("Feature Importance")
plt.show()
```

```python
#예측 모델 성능 향
df['views_log'] = np.log1p(df['view_count'])  # log(views + 1)
```

```python
#유튜브 데이터 시각화 인사이트

#시간대별 업로드 경향
sns.countplot(data=df, x='publish_hour')
plt.title("시간대별 업로드 빈도")
plt.xlabel("업로드 시간 (24시 기준)")
plt.ylabel("영상 수")
plt.show()
```

```python
#좋아요 수와 조회수의 관계
sns.scatterplot(data=df, x='likes', y='view_count')
plt.title("좋아요 수 vs 조회수")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
df_corr = df[(df['likes'] > 0) & (df['view_count'] > 0)]

# 피어슨 상관계수와 p-value 계산
corr, p_value = pearsonr(df_corr['likes'], df_corr['view_count'])

print(f"📈 피어슨 상관계수: {corr:.4f}")
print(f"📊 p-value: {p_value:.10f}")
```

```python
sns.regplot(x='view_count', y='likes', data=df)
plt.title('Likes vs View Count with Regression Line')
plt.show()
```

```python
df_corr['log_likes'] = np.log(df_corr['likes'])
df_corr['log_view_count'] = np.log(df_corr['view_count'])

# 로그 스케일에서도 상관관계 확인
corr_log, p_log = pearsonr(df_corr['log_likes'], df_corr['log_view_count'])
print(f"로그 스케일 상관계수: {corr_log:.4f}, p-value: {p_log:.10f}")
```
